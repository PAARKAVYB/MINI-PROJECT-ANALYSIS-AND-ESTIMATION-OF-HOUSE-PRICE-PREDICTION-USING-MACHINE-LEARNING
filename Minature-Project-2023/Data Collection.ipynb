#### Python Libaray for web scraping

```python
from selenium import webdriver

driver = webdriver.Chrome()
```

#### Connecting to website

```python
#driver.get("https://www.olx.in/")

driver.get("https://www.olx.in/for-sale-houses-apartments_c1725")
```

#### Loadmore till to end of page

```python
# loadmore

import time

while True:
    try:
        loadmore = driver.find_element_by_class_name('JbJAl')
        time.sleep(2)
        loadmore.click()
        time.sleep(5)
    except Exception as e:
        print(e)
        break

print("Completed Scrolling/Load More........")
#time.sleep(10)
#driver.quit()
```

#### Getting House ads links

```python
prodlist = driver.find_elements_by_class_name("EIR5N")
links = []
for prod in prodlist:
    try:
        urlele = prod.find_element_by_tag_name('a')
        links.append(urlele.get_property('href'))
    except:
        links.append("No Link")
        continue
```

```python
links[0:5]
```

This each link will contain the features along with price

```python
len(links)
```

#### Extracting Data from each link for the selected features

```python
import numpy as np

nulls = []
types = []
sqft = []
bedrooms = []
bathrooms = []
furnishing = []
construction = []
floor = []
facing = []
price = []
place = []

for i in range(0,len(links)):
    print(i)
    try:
        driver.get(links[i])
        
        try:
            price.append(driver.find_element_by_class_name("_2wMiF").find_element_by_class_name('_2xKfz').text)
        except:
            price.append(np.nan)
        try:
            place.append(driver.find_element_by_class_name("_2A3Wa").find_element_by_class_name('_2FRXm').text)
        except:
            place.append(np.nan)
            
        try:
            colval = driver.find_element_by_class_name('cb14b').find_element_by_class_name('_3JPEe').find_elements_by_class_name("_3_knn")
            colvaldict = dict()
            
            for i in colval:
                m = i.text.split("\n")[0]
                n = i.text.split("\n")[1:]
                colvaldict[m] = n
                
            #print(colvaldict)
            
            if 'Type' in colvaldict.keys():
                types.append(colvaldict['Type'])
            else:
                types.append(np.nan)
                
            if 'Bedrooms' in colvaldict.keys():
                bedrooms.append(colvaldict['Bedrooms'])
            else:
                bedrooms.append(np.nan) 
                
            if 'Bathrooms' in colvaldict.keys():
                bathrooms.append(colvaldict['Bathrooms'])
            else:
                bathrooms.append(np.nan) 
                
            if 'Furnishing' in colvaldict.keys():
                furnishing.append(colvaldict['Furnishing'])
            else:
                furnishing.append(np.nan) 
                
            if 'Construction Status' in colvaldict.keys():
                construction.append(colvaldict['Construction Status'])
            else:
                construction.append(np.nan) 
                
            if 'Super Builtup area (ft²)' in colvaldict.keys():
                sqft.append(colvaldict['Super Builtup area (ft²)'])
            else:
                sqft.append(np.nan) 
                
            if 'Floor No' in colvaldict.keys():
                floor.append(colvaldict['Floor No'])
            else:
                floor.append(np.nan)

            if 'Facing' in colvaldict.keys():
                facing.append(colvaldict['Facing'])
            else:
                facing.append(np.nan)
                
        except:
            types.append(np.nan)
            sqft.append(np.nan)
            bedrooms.append(np.nan)
            bathrooms.append(np.nan)
            furnishing.append(np.nan)
            construction.append(np.nan)
            floor.append(np.nan)
            facing.append(np.nan)
            
    except:
        nulls.append(i)
        continue
```
